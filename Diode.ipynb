Actually I am going to use Shockley equation without understanding. There is a term known as *volt-equivalent of temparature* $V_T$ is defined by 

$$V_T \equiv \frac{k_B T}{e}$$

Where two of them are known constant ($k_B = 1.3 \times 10^{-23} m^2 kg {s^{-2}}K^{-1}, q = 1.6 \times 10^{-19}$C). For every practical purpose we can consider T = 300K.

```python
#Constants:
kB = 1.3e-23; q = 1.6e-19
#Parameter
T = 300 #K

V_T = kB*T/q
print(V_T, "Volts")
```

## Shockley equation:
 
 $$\Large I_D = I_S( e^{\frac{V_D}{\eta V_T}} - 1)$$
 
Where **D** is for *Diode*. $I_D$ and $V_D$ are current and voltage across diode. $I_S$ is saturation current ( more accurately [reverse saturation current](https://en.wikipedia.org/wiki/Saturation_current)) of the diode. It is normaly order of $\mu A$. For simplicity I took $I_S = 1 \mu A$. $\eta$ is called *ideality factor* it is 1 for germanium and approximately 2 for silicon (info from Millman anf Halkias *Integrated eletronics*).

### 1. Forward biased

Current across the Diode in case of forward biased condition can be found using Shockley equation. I am using silicon diode s o $\eta = 1$.

```python
#Parameter
eta = 2
I_S = 1e-3 #mA

V_D = np.linspace(0, 0.7, 101)
I_D = I_S*(np.exp(V_D/(eta*V_T) - 1))

plt.figure(figsize = (14, 10))
plt.title("Forward biased", size = 30)
plt.plot(V_D, I_D)
plt.xlabel(r"$V_D$(volts)", size = 20)
plt.ylabel(r"$I_D$(mA)", size = 20)
plt.show()
```

```python
I_S = 100 #Micro A

V_D = np.linspace(-10, 0, 101)
I_D = I_S*(np.exp(V_D/(eta*V_T) - 1))

plt.figure(figsize = (14, 10))
plt.title("Reversed biased", size = 30)
plt.plot(V_D, I_D)
plt.xlabel(r"$V_D$(volts)", size = 20)
plt.ylabel(r"$I_D$(micro A)", size = 20)
plt.show()
```
